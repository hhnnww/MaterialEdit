```python
from pyzbar import pyzbar
from PIL import Image
from pathlib import Path

num = 1
for in_file in Path(r'D:\素材\8000-8999\8183\8183').iterdir():
    if in_file.is_file() and in_file.suffix.lower() in ['.png']:
        print(num)
        num += 1
        with open(in_file.as_posix(), 'rb') as f:
            im = Image.open(f).convert('L')

        if len(pyzbar.decode(im)) > 0:
            in_file.unlink()
```
