```python
from pathlib import Path
import shutil

material_path_list = [
    r'G:\饭桶设计\0-999', r'G:\饭桶设计\1000-1999'
]


def fun_包含文件(in_path: Path):
    for in_file in in_path.rglob('*'):
        if in_file.is_file():
            return True

    return False


for material_path in material_path_list:
    material_path = Path(material_path)

    for in_path in material_path.iterdir():
        if fun_包含文件(in_path) is False:
            print(in_path.as_posix())
            shutil.rmtree(in_path)
```
