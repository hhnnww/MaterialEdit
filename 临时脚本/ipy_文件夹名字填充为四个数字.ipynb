```python
from pathlib import Path

for in_file in Path(r'G:\泡泡素材\0000-0999').iterdir():
    if in_file.is_dir():
        if len(in_file.stem) < 4:

            ori_stem = in_file.stem
            while len(ori_stem) < 4:
                ori_stem = "0" + ori_stem

            new_path = in_file.with_stem(ori_stem)

            print(f'{in_file}\t-->\t{new_path}')
            in_file.rename(new_path)
```
